```python
import mass
from mass.io import sbml
from mass import Simulation

model = sbml.read_sbml_model('trp_CCM_manual_merge.xml')
```

```python
model.initial_conditions
```

```python
#unfix metabolites
model.metabolites.glc__D_e.fixed
```

```python
model.metabolites.ac_e.fixed
```

```python
model.metabolites.glc__D_e.fixed = False
model.metabolites.glc__D_e.fixed
```

```python
model.metabolites.ac_e.fixed = False
model.metabolites.ac_e.fixed
```

```python
#set initial conditions
model.metabolites.glc__D_e.initial_condition
```

```python
model.metabolites.glc__D_e.initial_condition = 23/180
model.metabolites.glc__D_e.initial_condition
```

```python
model.metabolites.ac_e.initial_condition
```

```python
model.metabolites.ac_e.initial_condition = 0
model.metabolites.ac_e.initial_condition
```

```python
# Simulate
sim = Simulation(reference_model=model, verbose=True)
```

```python
solutions = sim.simulate(model, time=(0,0.1))
solutions
```

```python
conc_sol, flux_sol = solutions

# List the first 5 points of concentration solutions
for metabolite, solution in conc_sol.items():
    print("{0}: {1}".format(metabolite, solution))
```
